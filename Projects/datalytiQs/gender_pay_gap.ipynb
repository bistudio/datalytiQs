```python
# rename columns

df_fpt.rename(columns = {'Description':'Job Categories', 'Code':'Job Category Code','(thousand)':'Female Part Time Job Count', 'Median':'Female Part Time % Median'}, inplace=True)

# replace non numerical values in columns

df_fpt['Female Part Time Job Count'].replace({'x':''},inplace=True)
df_fpt['Female Part Time Job Count'].replace({':':''},inplace=True)
df_fpt['Female Part Time Job Count'].replace({"''": 0}, inplace=True)

# convert no of jobs 000s column to integer

df_fpt['Female Part Time Job Count'] = pd.to_numeric(df_fpt['Female Part Time Job Count'], errors='coerce')

# convert % Median to float

df_fpt['Female Part Time % Median'] = pd.to_numeric(df_fpt['Female Part Time % Median'], errors='coerce')

df_fpt
```

```python
# merge full time data to previous data
final_output = df_1.merge(female_part_time, left_on='Job Category Code', right_on='Job Category Code')
final_output.drop(columns=['Job Categories'], axis=1, inplace=True)

# rename columns

final_output.rename(columns = {'Job Categories_x':'Job Categories'}, inplace=True)

final_output
```

```python
# rename columns

final_output.rename(columns = {'Male % Median':'Male Median Hourly Rate £', 'Female % Median':'Female Median Hourly Rate £', 'Full Time % Median':'Full Time Median Hourly Rate £'
                              , 'Part Time % Median':'Part Time Median Hourly Rate £', 'Male Full Time % Median':'Male Full Time Median Hourly Rate £', 'Male Part Time % Median':'Male Part Time Median Hourly Rate £'
                              , 'Female Full Time % Median':'Female Full Time Median Hourly Rate £' , 'Female Part Time % Median':'Female Part Time Median Hourly Rate £'}, inplace=True)

# create additional data points for visualization

final_output['Median Hourly £ Pay Gap'] = final_output['Male Median Hourly Rate £'] - final_output['Female Median Hourly Rate £']
final_output['Median Hourly % Pay Gap'] = 1 - (final_output['Female Median Hourly Rate £'] / final_output['Male Median Hourly Rate £'])
final_output['Full Time Median Hourly £ Pay Gap'] = final_output['Male Full Time Median Hourly Rate £'] - final_output['Female Full Time Median Hourly Rate £']
final_output['Full Time Median Hourly % Pay Gap'] = 1 - (final_output['Female Full Time Median Hourly Rate £'] / final_output['Male Full Time Median Hourly Rate £'])
final_output['Part Time Median Hourly £ Pay Gap'] = final_output['Male Part Time Median Hourly Rate £'] - final_output['Female Part Time Median Hourly Rate £']
final_output['Part Time Median Hourly % Pay Gap'] = 1 - (final_output['Female Part Time Median Hourly Rate £'] / final_output['Male Part Time Median Hourly Rate £'])

# add source / target img file paths

#df['<= 53'] = df['mynumbers'].apply(lambda x: 'True' if x <= 53 else 'False')

final_output
```

```python
# append dataframe

combined_df = pd.DataFrame({'workType':'All','jobCategoryCode':final_output['Job Category Code'], 'jobCategories':final_output['Job Categories'], 'maleJobCount':final_output['Male Job Count'], 'maleMedianHourlyRate':final_output['Male Median Hourly Rate £'], 'femaleJobCount':final_output['Female Job Count'],'femaleMedianHourlyRate':final_output['Female Median Hourly Rate £']
                            , 'medianPayGap':final_output['Median Hourly £ Pay Gap'], 'medianPayGapPct':final_output['Median Hourly % Pay Gap']})
combined_df

fullTime_df = pd.DataFrame({'workType':'Full Time','jobCategoryCode':final_output['Job Category Code'], 'jobCategories':final_output['Job Categories'], 'maleJobCount':final_output['Male Full Time Job Count'], 'maleMedianHourlyRate':final_output['Male Full Time Median Hourly Rate £'], 'femaleJobCount':final_output['Female Full Time Job Count'],'femaleMedianHourlyRate':final_output['Female Full Time Median Hourly Rate £']
                            , 'medianPayGap':final_output['Full Time Median Hourly £ Pay Gap'], 'medianPayGapPct':final_output['Full Time Median Hourly % Pay Gap']})

fullTime_df

partTime_df = pd.DataFrame({'workType':'Part Time','jobCategoryCode':final_output['Job Category Code'], 'jobCategories':final_output['Job Categories'], 'maleJobCount':final_output['Male Part Time Job Count'], 'maleMedianHourlyRate':final_output['Male Part Time Median Hourly Rate £'], 'femaleJobCount':final_output['Female Part Time Job Count'],'femaleMedianHourlyRate':final_output['Female Part Time Median Hourly Rate £']
                            , 'medianPayGap':final_output['Part Time Median Hourly £ Pay Gap'], 'medianPayGapPct':final_output['Part Time Median Hourly % Pay Gap']})

partTime_df


appended_df = [combined_df, fullTime_df, partTime_df]
final_union = pd.concat(appended_df)
final_union
final_union['maleImagePath'] = '/static/assets/img/businessman.png'
final_union['femaleImagePath'] = '/static/assets/img/business_woman.png'

# export output to directory

from json import loads, dumps

result = final_union.reset_index().to_json(r'app\data\gender_pay_gap.json', orient='records')
```
